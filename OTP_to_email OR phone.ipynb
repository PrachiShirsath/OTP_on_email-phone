```python
import random
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from email_validator import validate_email, EmailNotValidError
from twilio.rest import Client

# OTP Generator
def generate_otp() -> str:
    return str(random.randint(1000, 99999999))

# Email Validation
def is_valid_email(email: str) -> bool:
    try:
        validate_email(email)
        return True
    except EmailNotValidError:
        return False

# Send Email OTP
def send_email_otp(receiver_email: str, otp: int):

    sender_email = "shirsathmahesh28@gmail.com"
    sender_password = "virk lcnf lhiz ziez"

    msg = MIMEMultipart()
    msg["From"] = sender_email
    msg["To"] = receiver_email
    msg["Subject"] = "OTP Verification"

    msg.attach(MIMEText(f"Your OTP is: {otp}", "plain"))

    try:
        server = smtplib.SMTP("smtp.gmail.com", 587)
        server.starttls()
        server.login(sender_email, sender_password)
        server.send_message(msg)
        server.quit()
        print("✅ OTP sent to Email")
    except Exception as e:
        print("❌ Email Error:", e)

# Send SMS OTP (Twilio)
def send_sms_otp(phone: str, otp: int):
    account_sid = "ACXXXXX"
    auth_token = "XXXX"
    twilio_number = "+1XXXX"

    client = Client(account_sid, auth_token)

    if "+91" not in phone:
      phone = "+91" + phone

    try:
        message = client.messages.create(
            body=f"Your OTP is: {otp}",
            from_=twilio_number,
            to=phone
        )
        print("✅ OTP sent to Mobile")
    except Exception as e:
        print("❌ SMS Error:", e)



# -------- MAIN --------
email = input("Enter Email: ")
phone = input("Enter Mobile: ")


if not is_valid_email(email):
    print("❌ Invalid Email")
else:
    otp = generate_otp()
    print("Generated OTP:", otp)

    send_email_otp(email, otp)
    send_sms_otp(phone, otp)
```
